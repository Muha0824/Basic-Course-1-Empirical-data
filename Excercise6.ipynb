# Excercise 6.2
This is testing if i calculated by hand, correctly. This is excercise 6.2 and 2

```python
df = pd.DataFrame (
data = {
    # Its the Concentration of the compound (NH4)
    'Conc.': [0.0 , 1.0 , 5.0 , 10.0 , 20.0 , 30.0 , 40.0 , 50.0],

    # Its the Light absorbance (ABS) at a specific wavelength
    'ABS.': [0.083, 0.115, 0.191 , 0.299 , 0.501 , 0.695 , 0.893 , 1.084]
})

# Making a scatterplot
df.plot.scatter(
    figsize=(4,3),
    x = 'Conc.',
    y = 'ABS.'
)

x = df['Conc.']
y = df['ABS.']

variance = x.var()  # The variance of x
covariance = x.cov(y)   # The covariance between x and y

a = x.cov(y)/x.var()  # The slope
b = y.mean()-a*x.mean() # The intersection on y

# Fitting the curve
y_fit = a*x + b
df['fit'] = y_fit
plt.plot(x,y_fit, '-')

print(f"""The variance is {variance}, and the covariance is, {covariance}. 
      And the slope is {a} , and it intersects on y at {b}""")

# How to round the numbers, such that it's more accurate (2 decimals)? 
```

```python
# The same as before, but just in numpy
p = np.polyfit(x,y,1)
a,b = p

print(f'The slope is {a}')
print(f'The intersection is {b}')
```

# Exercise 6.3: Fit timber data

```python
# 1. Consider the problem where diameter plays the role of x and volume the role of y.
# Plot the data as a scatter plot.

df = pd.read_csv('timber.csv')  # reading the file

# Plotting it in scatterplot
plt.scatter(df['Diameter (in)'], df['Volume (cubic feet)'])
plt.xlabel('Diameter (in)')
plt.ylabel('Volume (cubic feet)')
plt.show()

# 2. Compute the slope a and the intersection b of the best fitting line.
x = df['Diameter (in)'] 
y = df['Volume (cubic feet)']

best_fit = np.polyfit(x,y,1)
a , b = best_fit
print(f"The slope of the best fit line is {a}, and the intersection is {b}")

# 3. Compute the values of the best fit for each of the x values (y_fit = a * x + b)
# and include them in the DataFrame (df) as a column named ' fit '.
y_fit = a*x+b
df['fit'] = y_fit

# check number 3 in the next cell
```

```python
df.head()   # 3 
```

```python
# 4. Make a scatter plot that includes the line of the best fit.
plt.scatter(df['Diameter (in)'], df['Volume (cubic feet)'])
plt.xlabel('Diameter (in)')
plt.ylabel('Volume (cubic feet)')
plt.plot(x,y_fit)
plt.show()
```

```python
# 4 Compute the residuals and add them in the DataFrame 
# as a column named residuals 

residuals = y-y_fit
df['residuals'] = residuals
df.head()
```

```python
# 6. Plot the residuals (use a stem plot).
plt.stem(x,residuals)
plt.xlabel('Diameter')
plt.ylabel('Volume')
```

# Exercise 6.4: Fit popcorn data

```python
# 1. Make a scatter plot of the short (x-axis)
#  versus the long (y-axis) dimensions of the popcorn.

df = pd.read_csv('popcorn_data.csv')    # Reading the file

# Defining the variables
x = df['short length']
y = df['long length ']

# Plotting it in scatter plot
plt.scatter(x,y)
plt.xlabel('Short length (mm)')
plt.ylabel('Long length (mm)')
plt.show()

# 2. Determine the slope a and the intersection for the best fit line with the short
# dimension as x and the long dimension as y
best_fit = np.polyfit(x,y,1)
a , b = best_fit
print(f"The slope of the best fit line is {a}, and the intersection is {b}")
```

# Exercise 6.5: Your fitting function*

```python
# 1. Use the function to fit the timber data of this chapter 

df = pd.read_csv('timber.csv')    # Reading the file

# Making a function for the timber data 
def best_linear_fit(x,y):
    # plotting the function
    plt.scatter(x,y)
    plt.xlabel('Diameter')
    plt.ylabel('Volume)')

    # calculating the residuals 
    p = np.polyfit(x,y,1)
    a,b = p 
    y_fit = a*x + b
    df['fit'] = y_fit
    residuals = y-y_fit
    df['residuals'] = residuals
    plt.stem(x,residuals)
    plt.show()

x = df['Diameter (in)']
y = df['Volume (cubic feet)']

best_linear_fit(x,y)
```
